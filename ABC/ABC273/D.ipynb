```python
H, W, Rs, Cs = map(int, input().split())
N = int(input())
RC = [list(map(int, input().split())) for _ in range(N)]
Q = int(input())
DL = [input().split() for _ in range(Q)]

LR = [[] for _ in range(H)]
LR

UD = [[] for _ in range(W)]
UD

for i in range(N):
    r, c = RC[i]
    LR[r-1].append(c-1)
    UD[c-1].append(r-1)

LR

UD

# from collections import defaultdict
# LR = defaultdict(int)
# UD = defaultdict(int)

# for i in range(N):
#     r, c = RC[i]
#     LR[r] = c
#     UD[c] = r

# dp = [[0]*W for _ in range(H)]
# dp

# for i in range(N):
#     r, c = RC[i]
#     dp[r-1][c-1] = 1
# dp

X, Y = Rs, Cs
for q in range(Q):
    d, l = DL[q]
    l = int(l)
    
    if d == "L" or d == "R":
        # Tと端の距離
        if d == "L":
            d1 = Y - 1
        elif d == "R":
            d1 = W - Y
        
        # Tと壁の距離
        d2 = float('inf')
        for kabe in LR[X-1]:
            d2 = min(d2, abs((Y-1)-kabe)-1)
            
    elif d == "U" or d == "D":
        # Tと端の距離
        if d == "U":
            d1 = X - 1
        elif d == "D":
            d1 = H - X
        
        # Tと壁の距離
        d2 = float('inf')
        for kabe in UD[Y-1]:
            d2 = min(d2, abs((X-1)-kabe)-1)
            
    D = min(d1, d2, l)
    
    # print (d, d1, d2, l)
    # print (D)
        
    if d == "L":
        Y -= D
    elif d == "R":
        Y += D
    elif d == "U":
        X -= D
    elif d == "D":
        X += D
    
    print (X, Y)
```
